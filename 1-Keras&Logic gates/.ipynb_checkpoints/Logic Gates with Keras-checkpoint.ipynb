```python
import tensorflow as tf
import pandas as pd
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from keras.utils.vis_utils import plot_model
from tensorflow.python.keras.callbacks import TensorBoard
from time import time
from datetime import datetime
import os
os.environ['PATH']
```

# AND Gate

```python
X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])
and_Y = np.array([[0],[0],[0],[1]])
```

```python

AND=Sequential()
AND.add(Dense(1, input_dim=2, activation='sigmoid'))
AND.compile(loss='binary_crossentropy', optimizer='adam')#min kac tane node toplam olarak
#ve kac katman
#loss funct
#accuracy'i arastir iyice!!!

#AND icin acc yok, loos'u sifir yapmak lazim 
#loss ==0.02 oldugunda egitimi bitir diye emir verebiliriz
#optimizer'de lr'yi farkli degerleri ver ve dene
AND.summary()
```

```python
history = AND.fit(X, and_Y, epochs=1500, verbose=1)
```

```python
predictions=AND.predict(X)

predictions
```

```python
k=0
for i in predictions:
    
    if( i>= 0.5):
        print("input: ", X[k],"=" ,1)
    else:
        
        print("input: ", X[k],"=" ,0)
    k+=1
```

# OR

```python
or_Y = np.array([[0],[1],[1],[1]])

OR=Sequential()
OR.add(Dense(2, input_dim=2, activation='relu'))
OR.add(Dense(1, activation='sigmoid'))

OR.compile(loss='binary_crossentropy', optimizer='sgd')
print(OR.summary())
history = OR.fit(X, or_Y, epochs=1000)
```

```python
#predict
predictions=OR.predict(X)
k=0
for i in predictions:
    
    if( i> 0.5):
        print("input: ", X[k],"=" ,1)
    else:
        
        print("input: ", X[k],"=" ,0)
    k+=1
```

# XOR

```python
xor_Y = np.array([[0],[1],[1],[0]])

XOR=Sequential()
XOR.add(Dense(4, input_dim=2, activation='relu'))
XOR.add(Dense(1, activation='sigmoid'))
XOR.compile(loss='binary_crossentropy', optimizer='adam')#min kac tane node toplam olarak
print(XOR.summary())
history = XOR.fit(X, xor_Y, epochs=700)
```

```python
predictions=XOR.predict(X)
predictions
```

```python
#predict

k=0
for i in predictions:
    
    if( i> 0.5):
        print("input: ", X[k],"=" ,1)
    else:
        
        print("input: ", X[k],"=" ,0)
    k+=1
```

# XNOR

```python
xNor_Y = np.array([[1],[0],[0],[1]])

XNOR=Sequential()
XNOR.add(Dense(4, input_dim=2, activation='relu'))
XNOR.add(Dense(1, activation='sigmoid'))
XNOR.compile(loss='binary_crossentropy', optimizer='adam',  metrics=['acc'])
print(XNOR.summary())
history = XNOR.fit(X, xNor_Y, epochs=1500)
```

```python
#predict
predictions=XNOR.predict(X)
k=0
for i in predictions:
    
    if( i> 0.5):
        print("input: ", X[k],"=" ,1)
    else:
        
        print("input: ", X[k],"=" ,0)
    k+=1
```
